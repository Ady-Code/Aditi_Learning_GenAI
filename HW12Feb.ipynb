```python
books = {}
users = {}
borrow_history = []
books_category = set()

is_continue = True
menu_options = "Choose the option from the menu:\n1. Add Book\n2. Register User\n3. Borrow Book\n4. Return Book\n5. Show Library Analytics\n6. Memory Inspection Report\n7. Exit "

def add_book():
    try:

        book_id = input("Enter book id")

        if book_id in books:
            print("book ID already exist")
        title = input("Enter title")
        author = input("Enter author")
        category = input("Enter category")
        copies = int(input("Enter number of copies"))

        books[book_id] = {"title":title,
                        "author": author,
                        "category": category,
                        "copies": copies}
        
        print(books)
        books_category.add("Non fiction")
    except Exception as e:
        print(f"Error: {e}")

def register_user():
    try:
        user_id = input("Enter user ID")

        if user_id in users:
            print("User already exists")

        name = input("Enter name")

        users[user_id] = { "name": name,
                        "borrowed_books": []
                        }

        print("Registration successful")

    except Exception as e:
        print(f"Error: {e}")


# 3ï¸âƒ£BORROW BOOK

def borrow_book():
    try:

        user_id = input("Enter user ID: ")
        book_id = input("Enter book ID: ")

        if user_id not in users:
            print("user does not exist")

        if book_id not in books:
            print("book does not exist")

        if books[book_id]["copies"] == 0:
            print("No copies available!")

        books[book_id]["copies"] -= 1
        users[user_id]["borrowed_books"].append(book_id)
        borrow_history.append((user_id, book_id))

        print("Book borrowed successfully")

    except Exception as e:
        print(f"Error: {e}")


# 4ï¸âƒ£ RETURN BOOK

def return_book():
    try:

        user_id = input("Enter User ID: ")
        book_id = input("Enter Book ID: ")

        if user_id in users and book_id in users[user_id]["borrowed_books"]:
            users[user_id]["borrowed_books"].remove(book_id)
            books[book_id]["copies"] += 1
            print("Book return successful")
        else:
            print("Invalid return")

    except Exception as e:
        print(f"Error: {e}")

# 5ï¸âƒ£ ANALYTICS

def show_library_analytics():
    try:
        total_books = 0
        for book_id in books:
            total_books += books[book_id]["copies"]

        print("Total books in library:", total_books)
        print("Total unique categories:", len(books_category))

        borrow_count = {}

        for record in borrow_history:
            book_id = record[1]
            if book_id in borrow_count:
                borrow_count[book_id] += 1
            else:
                borrow_count[book_id] = 1

        # Most borrowed book
        most_borrowed = None
        max_count = 0

        for book_id in borrow_count:
            if borrow_count[book_id] > max_count:
                max_count = borrow_count[book_id]
                most_borrowed = book_id

        print("Most borrowed book:", most_borrowed)

        # User who borrowed most books
        top_user = None
        max_books = 0

        for user_id in users:
            count = len(users[user_id]["borrowed_books"])
            if count > max_books:
                max_books = count
                top_user = user_id

        print("User who borrowed most books:", top_user)

        # Books currently unavailable
        print("Books currently unavailable:")
        for book_id in books:
            if books[book_id]["available_copies"] == 0:
                print(book_id, "-", books[book_id]["title"])

    except Exception as e:
        print(f"Error: {e}")

# 6ï¸âƒ£ MEMORY INSPECTION

def memory_inspection_report():
    try:
        
        # A. Show memory id
        print("ID of books dictionary:", id(books))

        if len(books) > 0:
            first_book = list(books.keys())[0]
            print("one specific book entry id:", id(books[first_book]))

        if len(users) > 0:
            first_user = list(users.keys())[0]
            print("ID of borrowed_books list:",
                id(users[first_user]["borrowed_books"]))

        # B. Demonstrate reference behavior
        temp_books = books

        print("books id:", id(books))
        print("temp books id:", id(temp_books))

        if len(books) > 0:
            first_book = list(books.keys())[0]
            temp_books[first_book]["available_copies"] += 1
            print("Modified temp_books -> original books changed!")

        # C. Demonstrate shallow copy
        copied_books = books.copy()

        print("books id:", id(books))
        print("copied books id:", id(copied_books))

        if len(books) > 0:
            first_book = list(books.keys())[0]

            print("Nested ID original:",
                id(books[first_book]))
            print("Nested ID copied:",
                id(copied_books[first_book]))

            copied_books[first_book]["available_copies"] += 1
            print("Modified nested dictionary in copied_books")

        # D. Demonstrate deletion

        if len(users) > 0:
            delete_user = list(users.keys())[0]

            print("ID before deletion:", id(users))
            del users[delete_user]
            print("ID after deletion:", id(users))
        else:
            print("No users")

    except Exception as e:
        print(f"Error: {e}")

def menu():
    try: 

        while is_continue:
            print(menu_options)

            choice = input("Enter Choice:")
            if choice == "1":
                add_book()
            elif choice == "2":
                register_user()
            elif choice == "3":
                borrow_book()
            elif choice == "4":
                return_book()
            elif choice == "5":
                show_library_analytics()
            elif choice == "6":
                memory_inspection_report()
            elif choice == "7":
                print("Exit")
                # is_continue = False
                break
            else:
                print("Invalid choice!")
    except Exception as e:
        print(f"Error: {e}")
        
menu()

```
