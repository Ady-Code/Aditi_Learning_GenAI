Number Intelligence Analyzer

Inputs
	1.	A positive integer n

⸻

Tasks (must use a for loop)

For numbers from 1 to n, calculate:
	•	Sum of all numbers
	•	Sum of even numbers
	•	Sum of odd numbers
	•	Count of numbers divisible by 3

Output: 
Analysis from 1 to 20
--------------------
Total Sum: 210
Even Numbers Sum: 110
Odd Numbers Sum: 100
Divisible by 3 Count: 6

```python
any_num = int(input("Enter any +ve number"))
x = 0
y = 0
z = 0
p = 0

print(f"Analysis from 1 to {any_num}")
for i in range(1,any_num+1):
    x = i + x
    if (i%2==0):
        y = i + y
    else:
        z = i + z
    if (i%3==0):
        p +=1

print(f"Total Sum: {x}")
print(f"Even Numbers Sum: {y}")
print(f"Odd Numbers Sum: {z}")
print(f"Divisible by 3 Count: {p}")


```
