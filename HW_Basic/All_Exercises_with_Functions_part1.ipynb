```python
# Level 4: The Variable Swap (Logic Puzzle)
# glass_a = "Water"
# glass_b = "Milk"

def swap(glass_a,glass_b):
    print(f"Before: A={glass_a}, B={glass_b}")
    glass_a,glass_b = glass_b,glass_a
    print(f"After:  A={glass_a}, B={glass_b}")

swap("Water","Milk")
```

```python
# Level 3: The Pizza Splitter (Floor Division) ğŸ• 
# # Level 4: The Leftovers (Modulo) 

def pizza_slicer(slices,frnds):
    print(f"Each person gets {slices//frnds} slices.")
    print(f"The host gets the remaining {slices%frnds} slices.")

pizza_slicer(46,8)
```

```python
# a 3digit number is given, print ones, tens, hundreds

def number(num):
    print(f"Hundreds digit will be {num//100}")
    print(f"Tens digit will be {(num//10)%10}")
    print(f"Ones digit will be {num%10}")

number(472)
```
