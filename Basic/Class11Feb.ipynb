```python
a = [1, 2, 3]
b = a
# a = b

print(id(a), id(b))   # Same ID â†’ both point to same list

b.append(4)
print(a)
print(id(a), id(b))
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
# b = a
# a = b

print(id(a), id(b)) # stack memory
print(id(a[0]), id(b[0])) # heap memory

b.append(4)
print(a)
print(b)
print(id(a), id(b))
a.append(5)
print(a)
print(id(a), id(b))
```
