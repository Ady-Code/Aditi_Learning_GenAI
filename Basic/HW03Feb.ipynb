```python
name = str(input("Enter your name"))
age = int(input("Enter your age"))
# maths = int(input("Enter marks in Mathematics, out of 100"))
# sci = int(input("Enter marks in Science, out of 100"))
# eng = int(input("Enter marks in English, out of 100"))
# family_ann_inc = float(input("Enter you Family annual income"))
# res_category = str(input("Are you from a reserved category: Yes or No)"))

if age < 17:
    print(f"Invalid data: Come back in {17-age} years")
else:
    maths = int(input("Enter marks in Mathematics, out of 100"))
    sci = int(input("Enter marks in Science, out of 100"))
    eng = int(input("Enter marks in English, out of 100"))
    if (maths < 0) or (sci < 0) or (eng < 0):
        print("Invalid student data. Please check inputs.")
    elif (100 < maths) or (100 < sci) or (100 < eng):
        print("Invalid student data. Please check inputs.")
    else:
        total_marks = maths + sci + eng
        percentage = (total_marks / 300) * 100
        # print(f"Student name: {name}")
        # print(f"Percentage: {percentage}")
        if percentage >= 70 and sci >= 75 and maths >= 75:
            # print("You are eligible for Science Stream")
            stream = 'Science'
            base_fee = 120000.00
        elif percentage >= 60 and maths >= 60:
            # print("You are eligible for Commerce Stream")
            stream = 'Commerce'
            base_fee = 90000.00
        elif percentage >= 50:
            # print("You are eligible for Arts Stream")
            stream = 'Arts'
            base_fee = 60000.00
        else:
            print("Not eligible for admission")
        family_ann_inc = float(input("Enter you Family annual income"))
        if family_ann_inc <= 500000:
            discount = 30
            # print(discount)
        elif family_ann_inc <= 800000:
            discount = 20
            # print(discount)
        else:
            discount = 0
            # print(discount)
        res_category = bool(input("Are you from a reserved category: True or False)"))
        if res_category == 'True':
            discount +=10
        
        print("Admission Result")
        print("----------------")
        print(f"Name: {name}")
        print(f"Age: {age}")
        print(f"Stream Allotted: {stream}")
        print(f"Percentage: {percentage:.2f}%")
        # ask round function
        print("     ")
        
        print(f"Base Fee: {base_fee:.2f}")
        print(f"Total Discount: {discount}%")
        print(f"Final Fee to Pay: {base_fee*(1-(discount/100)):.2f}")


    
```
