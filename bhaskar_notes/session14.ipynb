## Scenario: E-commerce Order Processing System

You run an online store.
You have orders. You want to:

* Filter valid orders
* Apply discounts
* Calculate totals

We’ll use:

* `lambda`
* `map`
* `filter`
* `reduce`




```python
# Step 0 — Sample Data

orders = [
    {"id": 1, "customer": "Amit", "amount": 1200, "status": "completed"},
    {"id": 2, "customer": "Neha", "amount": 800, "status": "cancelled"},
    {"id": 3, "customer": "Raj", "amount": 1500, "status": "completed"},
    {"id": 4, "customer": "Sara", "amount": 400, "status": "completed"},
]
```

```python
def get_completed_order(order):
    if order["status"] == "completed":
        return True
    else:
        return False
```

```python
# filter() — Keep Only Valid Orders
# Only “completed” orders should be processed.
valid_orders = list(filter(get_completed_order, orders))
print(valid_orders)
```

```python
# filter() — Keep Only Valid Orders
# Only “completed” orders should be processed.
valid_orders = list(filter(lambda order: order["status"] == "completed", orders))
print(valid_orders)
```

### What happened?

* Each order goes through the lambda
* If True → kept
* If False → removed


```python
# map() — Apply Discount
# Apply 10% discount to all valid orders.

def apply_discount(order):
    order["amount"] = order["amount"] * 0.9
    return order

discounted_orders = list(
    map(apply_discount, valid_orders)
)
print(discounted_orders)
```

```python
# map() — Apply Discount
# Apply 10% discount to all valid orders.

discounted_orders = list(
    map(lambda order: {**order, "amount": order["amount"] * 0.9}, valid_orders)
)
print(discounted_orders)
```

```python
def add_amount(acc, order):
    return acc + order["amount"]
```

```python
# reduce() — Calculate Total Revenue
# total revenue after discount

from functools import reduce

total_revenue = reduce(
    add_amount,
    discounted_orders,
    0
)

print(total_revenue)
```
