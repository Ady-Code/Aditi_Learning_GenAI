```python
import csv
import json
import os

# 1. Create a dummy CSV file
with open(os.path.join("fileHandling", "students.csv"), "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Age", "Grade"]) # Header
    writer.writerow(["Tony", "15", "A"])
    writer.writerow(["Steve", "16", "B"])
    writer.writerow(["Natasha", "15", "A+"])

# 2. Create a dummy JSON file
data = {
    "school": "Midtown High",
    "location": "Queens",
    "is_open": True,
    "students": ["Peter", "MJ", "Ned"]
}
with open(os.path.join("fileHandling","school_config.json"), "w", newline="") as file:
    json.dump(data, file, indent=4) # indent makes it look pretty

print("✅ 'students.csv' and 'school_config.json' created!")
```

```python
# Reading a csv
import csv

print("--- Reading CSV ---")
with open(os.path.join("fileHandling", "students.csv"), "r", newline="") as file:
    reader = csv.reader(file) # Create a reader machine
    
    # Loop through the reader to get rows
    for row in reader:
        print(row) # Each row is a List of Strings
```

```python
# Writing/Appending to csv
new_student = ["Bruce", "17", "A"]

with open(os.path.join("fileHandling","students.csv"), "a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(new_student)

print("Added Bruce to the CSV.")
```

```python
import json

# 1. Reading JSON
print("--- Reading JSON ---")
with open(os.path.join("fileHandling","school_config.json"), "r", newline="") as file:
    school_data = json.load(file) # Converts text -> Dictionary

print(f"School Name: {school_data['school']}")
print(f"Student 1: {school_data['students'][0]}")

# 2. Modifying & Saving JSON
school_data["principal"] = "Mr. Morita" # Add new key

with open(os.path.join("fileHandling","school_config.json"), "w", newline="") as file:
    json.dump(school_data, file, indent=4) # Save it back

print("JSON updated.")
```
