## Working with modules - Feb23
- packages/modules/libraries - all are same
- packages
    - inbuilt modules
    - external modules
- use `import` keyword to import packages
- use `from package_name import function_name` to import a specific function from a package
- to create an alias, use `as` keyword- `import datetime as dt`
- ideal way to use pip is to create virtual environment- `python -m venv virtual_env_name(.venv)`
- to activate the virtual env, run command `.\.venv\Scripts\Activate.ps1`
    - for unix machine, run command `source .venv/bin/activate`
- to install a package, run command `pip instal package_name`
    - to install packages from requirements.txt, run command  `pip install -r requirements.txt(path of requirements file)`
- to create requiremets.txt, run command `pip freeze > requirements.txt(path of requirements files)`
