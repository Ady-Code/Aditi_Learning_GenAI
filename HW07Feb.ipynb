Phase 1: Login System (While Loop)
Rules

	•	User gets maximum 3 attempts

Behaviour
	•	If correct PIN → login successful
	•	If wrong PIN → show attempts left
	•	After 3 failures → card blocked, program exits

Output example

Once logged in, repeatedly show this menu:

Phase 4: Expense Tracker (For + While Combined)

This is the most important part.

Behavior
	•	Ask user how many expenses they want to record (max 5)
	•	Use a for loop to collect expenses
	•	Each expense is just a number (amount)

While recording:
	•	Ignore negative values (use continue)
	•	Accumulate total expenses
	•	Track:
	•	highest expense
	•	lowest expense

```python
# Phase 1: Login System (While Loop)
is_continue = True
pin = 4321
attempt = 1
menu_options = "Choose the option from the ATM menu:\n1. Check Balance\n2. Deposit Money\n3. Withdraw Money\n4. Expense Tracker\n5 Exit"
is_menu = True
balance = 10000
total_expense = 0

while is_continue:
    guess = int(input("Enter the pin"))
    if pin != guess:
        if attempt < 3:
            print(f"Try guessing the pin again, you have {3 - attempt} attempts left")
            attempt+=1
            continue
        else:
            print(f"you have 0 attempts left,come back in 30min")
            is_continue = False
            # break
    else:
        print("login successful")
# Phase 2: ATM Menu (Infinite While Loop)
# Phase 3: Core Banking Operations
        
                
        while is_menu:
            print(menu_options)
            menu = int(input("Choose the option from the ATM menu:\n1. Check Balance\n2. Deposit Money\n3. Withdraw Money\n4. Expense Tracker\n5 Exit"))

            if menu == 5:
                print("You choose to exit","Thank you for using the ATM","Session ended safely.",sep="\n")
                is_menu = False
                # break
            elif menu == 1:
                print(f"Current Balance:{balance}")
                continue
            elif menu == 2: #Deposit Money
                dep_amount = int(input("Enter deposit amount"))
                if dep_amount > 0:
                    balance += dep_amount
                    print(f"₹{dep_amount} deposited successfully",f"Updated Balance: {balance}",sep="\n")
                    continue
                else:
                    print("Enter a valid amount")
                    continue
            elif menu == 3: #Withdraw Money
                withdraw = int(input("Enter withdraw amount"))
                if (withdraw > 0) and (withdraw <= balance):
                    balance -= withdraw
                    print(f"₹{withdraw} withdrawn successfully",f"Updated Balance: {balance}",sep="\n")
                    continue
                else:
                    print("Insufficient balance or invalid amount")
                    continue
            elif menu == 4: #Expense Tracker
                # expense = int(input("Enter expenses you want to record"))
                max_expense = 1
                min_expense = balance
                for i in range(0,5):
                    expense = int(input("Enter expenses you want to record"))
                    total_expense += expense
                    i += 1
                    if max_expense < expense:
                        max_expense = expense
                    else:
                        max_expense = max_expense
                    if min_expense < expense:
                        min_expense = min_expense
                    else:
                        min_expense = expense

                balance -= total_expense
                print(f"Total Expenses: {total_expense}")
                print(f"Highest Expense: {max_expense}")
                print(f"Lowest Expense: {min_expense}")
                print(f"Remaining Balance: {balance}")
                    

            else:
                continue
        is_continue = False
        # break
```
