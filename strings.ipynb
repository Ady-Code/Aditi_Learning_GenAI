```python
movies = ["Batman", "Superman", "Spiderman"]

# 1. SWAP: Changing an item
movies[1] = "Wonder Woman" 
print("After Swap:", movies)

# 2. APPEND: Adding to the END (Most common!)
movies.append("Flash")
print("After Append:", movies)

# 3. INSERT: Adding to a specific spot
movies.insert(0, "Aquaman") # Puts Aquaman at the very front
print("After Insert:", movies)

print(id(movies))
# 4. Append using cancatination
movies = movies + ["Ant-Man","ww"] #unpacking
print(f"After Concatination: {movies}")
print(id(movies))

```

```python
hero = {
    "name": "Bruce Wayne",
    "alias": "Batman",
    "age": 40,
    "is_rich": True
}
print(hero["name"])  # Bruce Wayne
print(hero["alias"]) # Batman

# ❌ This will CRASH if the key doesn't exist
# print(hero["power_level"]) # KeyError!

# ✅ Safer Way: .get()
# If the key is missing, it returns 'None' instead of crashing.
print(hero.get("power_level",40))
print(hero)
```
