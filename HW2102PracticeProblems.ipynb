```python
# Problem4: Nested Function Challenge

numbers = []

def process_numbers(numbers):
    def even_numbers():
        even_num = [i for i in numbers if i % 2 == 0]
        print(f"Evens: {even_num}")
    def avg_numbers():
        try:
            print(f"Avg: {sum(numbers)/len(numbers)}")

        except ZeroDivisionError:
            print("Error occurred: division by zero")
    
    even_numbers()
    avg_numbers()
    
process_numbers([1,2,4,6,10,10,20])
process_numbers([])

    
```
