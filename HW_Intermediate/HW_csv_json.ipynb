```python
# Level 1: The Inventory Maker (CSV Write) üì¶
# Create a list of lists: inventory = [["Item", "Count"], ["Apple", 10], ["Banana", 5]].
# Open inventory.csv in Write ("w") mode.
# Use csv.writer to write all rows.

import os
import csv

inventory_list = [["Item", "Count"], ["Apple", 10], ["Banana", 5]]
with open(os.path.join(r"C:\\Users\\Admin\\Desktop\\projects\Aditi_Learning_GenAI\\fileHandling","inventory.csv"),"w", newline="") as inventory:
    inventoryRows = csv.writer(inventory)
    for i in inventory_list:
        inventoryRows.writerow(i)
```
