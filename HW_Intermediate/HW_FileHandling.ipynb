```python
# Level 2: The Auto-Diary üìî
# Refactor Level 1 using the with statement.

# Ask the user "How are you feeling?".
# Save their answer to diary.txt.
# Print "Saved."

with open("test.txt","a") as diary:
    user_input = input("How are you feeling today?")
    diary.write("\n"+user_input)
print("User Input saved to the file")
```

```python
# Level 3: The Guestbook (Append) üñäÔ∏è
# Use a while True loop.
# Ask for a guest name. (Type "exit" to stop).
# Open guestbook.txt in Append mode.
# Write the name + a newline character \n.

is_continue = True

while is_continue:
    guest = input("Enter your name or type 'exit' to stop")
    if guest == 'exit':
        break
    with open("guestbook.txt",'a') as guestbook:
        guestbook.write(f"{guest}\n")
    is_continue = False
    
```

```python
# Level 4: The Line Counter üî¢
# Manual Setup: Create a file named story.txt manually (Right click -> New File) and paste 4-5 lines of text in it. Task:

# Write a Python script to read story.txt.
# Count how many lines are in the file.
# Print: "The story has X lines."

with open("story.txt","r+") as story:
    print(f"The story has {len(story.readlines())} lines")
```

```python
# Level 5: The "Error" Logger (Logic + Files) ‚ö†Ô∏è
# You have a list of server logs: logs = ["INFO: System OK", "ERROR: Database Fail", "INFO: User login", "ERROR: WiFi down"]

# Task:

# Loop through the list.
# If the log starts with "ERROR", write it to a file called errors_only.txt.
# Ignore the "INFO" logs.
# Read back errors_only.txt to verify.

logs = ["INFO: System OK", "ERROR: Database Fail", "INFO: User login", "ERROR: WiFi down"]

for i in logs:
    if i.startswith("ERROR"):
        with open("errors_only.txt","a") as Errors:
            Errors.write(i+"\n")
        print(i)
```
