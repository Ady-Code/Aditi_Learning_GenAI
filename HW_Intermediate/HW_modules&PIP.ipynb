```python
# Level 4: The Fun Library (PIP) üé®
# Use !pip install colorama (if not installed).
# Import Fore, Back, Style from colorama.
# Print text in RED.
# Print text with a GREEN background.
# Reset the style at the end using Style.RESET_ALL.

from colorama import Fore, Back, Style
print(Fore.RED + "This text is red")
print(Back.GREEN + "This text is green")
print(Style.BRIGHT + "This text is bright")
print(Style.RESET_ALL + "Back to normal text")
print(Style.BRIGHT + Fore.RED + Back.GREEN + "Bright red text with green background")
print(Style.RESET_ALL) # reset first before applying any other style
print(Style.DIM + "This will be dim text")
print(Style.NORMAL + "Is this normal text?")
```
